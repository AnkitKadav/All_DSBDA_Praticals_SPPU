```python
data = pd.read_csv('iris.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize = (16, 9))
sns.histplot(data['sepallength'], ax = axes[0, 0])
sns.histplot(data['sepalwidth'], ax = axes[0, 1])
sns.histplot(data['petallength'], ax = axes[1, 0])
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize = (16, 9))
sns.boxplot(y = 'petallength', x = 'class', data = data, ax =
axes[0,0])
sns.boxplot(y = 'petalwidth', x = 'class', data = data, ax =
axes[0,1])
sns.boxplot(y = 'sepallength', x = 'class', data = data, ax =
axes[1,0])
sns.boxplot(y = 'sepalwidth', x = 'class', data = data, ax =
axes[1,1])
```
